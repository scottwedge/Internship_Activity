```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import csv
import pandas as pd
import re

# recent 긁고, 날짜 필터링하고, upcoming과 recent로 분류
# upcoming 긁고, 날짜 필터링하고, upcoming과 recent로 분류
# 모든 파일에서 중복 제거




driver = webdriver.Chrome('../../chromedriver/chromedriver')
url = 'https://www.themoviedb.org/movie/now-playing'
driver.get(url)
    
start = time.time()
    
driver.maximize_window()
    
time.sleep(2)
    
# 필터 설정 시작

driver.find_element_by_xpath('//*[@id="media_v4"]/div/div/div[2]/div[1]/div[1]/div[2]/span/span/span[2]/span')
time.sleep(1)
driver.find_element_by_xpath('')
    
driver.find_element_by_xpath('//*[@id="media_v4"]/div/div/div[2]/div[1]/div[2]/div[1]').click()
time.sleep(1)
driver.find_element_by_xpath('//*[@id="all_release_countries"]').click()
time.sleep(1)
driver.find_element_by_xpath('//*[@id="search_all_countries"]/span/span/span[2]/span').click()
time.sleep(1)

box = driver.find_element_by_xpath('//*[@id="release_region-list"]/span/input')
time.sleep(1)
box.clear()
time.sleep(1)
box.send_keys("United States")
time.sleep(1)
driver.find_element_by_xpath('//*[@id="release_region_listbox"]/li[2]').click()
time.sleep(1)

driver.find_element_by_xpath('//*[@id="media_v4"]/div/div/div[2]/div[3]/p/a').click()

time.sleep(5)

driver.find_element_by_xpath('//*[@id="media_v4"]/div/div/div[2]/div[1]/div[3]/div[1]').click()
time.sleep(1)
driver.find_element_by_xpath('//*[@id="media_v4"]/div/div/div[2]/div[1]/div[3]/div[3]/span[1]/span/span[2]/span').click()
time.sleep(1)

box2 = driver.find_element_by_xpath('//*[@id="ott_region-list"]/span/input')
time.sleep(1)
box2.clear()
time.sleep(1)
box2.send_keys("United States")
time.sleep(1)
box2.send_keys(Keys.RETURN)
time.sleep(1)

driver.find_element_by_xpath('//*[@id="media_v4"]/div/div/div[2]/div[3]/p/a').click()

time.sleep(5)
    
# 필터 설정 끝
    
for i in range(0, 5) :
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")  
    time.sleep(1)
        
    try :
        driver.find_element_by_css_selector('#pagination_page_1 > p > a').click()
    except :
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
quest_list = driver.find_elements_by_css_selector('div > div.image > div.wrapper > a')
quest_urls = [ i.get_attribute('href') for i in quest_list ]
print(len(quest_urls))
    
temp_id = []
temp_title = []
temp_release = []
temp_genre = []
temp_director = []
#temp_actor = []
temp_country = []
temp_runtime = []    
temp_overview = []
temp_url = []
    
if quest_urls != [] :
    for quest in quest_urls :
        driver.get(quest)
        time.sleep(0.5)
            
        #URL 주소
        url = quest
            
        # Site_ID, 정규표현식사용
        site_id = re.findall("\d+", url)
        site_id = site_id[0]
            
        # 영화 제목
        title = driver.find_elements_by_css_selector('#original_header > div > section > div > h2 > a')[0].text
            
        # 개봉일, 정규표현식사용
        release = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper > section > div.title > h2 > span')[0].text
        release = release.replace("(", "")
        release = release.replace(")", "")
            
        try :
            try : 
                a = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.false > section > div.header_info > ol > li:nth-child(1) > p.character')[0].text
                b = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.false > section > div.header_info > ol > li:nth-child(2) > p.character')[0].text
                c = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.false > section > div.header_info > ol > li:nth-child(3) > p.character')[0].text
                d = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.false > section > div.header_info > ol > li:nth-child(4) > p.character')[0].text  
                e = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.false > section > div.header_info > ol > li:nth-child(5) > p.character')[0].text
                f = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.false > section > div.header_info > ol > li:nth-child(6) > p.character')[0].text
                g = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.false > section > div.header_info > ol > li:nth-child(7) > p.character')[0].text
                h = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.false > section > div.header_info > ol > li:nth-child(8) > p.character')[0].text
            except :    
                a = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.true > section > div.header_info > ol > li:nth-child(1) > p.character')[0].text
                b = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.true > section > div.header_info > ol > li:nth-child(2) > p.character')[0].text
                c = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.true > section > div.header_info > ol > li:nth-child(3) > p.character')[0].text
                d = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.true > section > div.header_info > ol > li:nth-child(4) > p.character')[0].text
                e = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.true > section > div.header_info > ol > li:nth-child(5) > p.character')[0].text
                f = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.true > section > div.header_info > ol > li:nth-child(6) > p.character')[0].text
                g = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.true > section > div.header_info > ol > li:nth-child(7) > p.character')[0].text
                h = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.true > section > div.header_info > ol > li:nth-child(8) > p.character')[0].text
        except : 
            pass
            
        # 감독
            
        try :
            if 'Director' in a :
                try :
                    director = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.false > section > div.header_info > ol > li:nth-child(1) > p:nth-child(1) > a')[0].text
                except :
                    director = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.true > section > div.header_info > ol > li:nth-child(1) > p:nth-child(1) > a')[0].text
            elif 'Director' in b :
                try :
                    director = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.false > section > div.header_info > ol > li:nth-child(2) > p:nth-child(1) > a')[0].text
                except :
                    director = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.true > section > div.header_info > ol > li:nth-child(2) > p:nth-child(1) > a')[0].text
            elif 'Director' in c :   
                try :
                    director = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.false > section > div.header_info > ol > li:nth-child(3) > p:nth-child(1) > a')[0].text
                except :
                    director = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.true > section > div.header_info > ol > li:nth-child(3) > p:nth-child(1) > a')[0].text
            elif 'Director' in d :
                try :
                    director = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.false > section > div.header_info > ol > li:nth-child(4) > p:nth-child(1) > a')[0].text
                except :
                    director = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.true > section > div.header_info > ol > li:nth-child(4) > p:nth-child(1) > a')[0].text
            elif 'Director' in e :
                try :
                    director = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.false > section > div.header_info > ol > li:nth-child(5) > p:nth-child(1) > a')[0].text
                except :
                    director = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.true > section > div.header_info > ol > li:nth-child(5) > p:nth-child(1) > a')[0].text
            elif 'Director' in f :   
                try :
                    director = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.false > section > div.header_info > ol > li:nth-child(6) > p:nth-child(1) > a')[0].text
                except :
                    director = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.true > section > div.header_info > ol > li:nth-child(6) > p:nth-child(1) > a')[0].text
            elif 'Director' in g :
                try :
                    director = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.false > section > div.header_info > ol > li:nth-child(7) > p:nth-child(1) > a')[0].text
                except :
                    director = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.true > section > div.header_info > ol > li:nth-child(7) > p:nth-child(1) > a')[0].text
            elif 'Director' in h :
                try :
                    director = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.false > section > div.header_info > ol > li:nth-child(8) > p:nth-child(1) > a')[0].text
                except :
                    director = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.true > section > div.header_info > ol > li:nth-child(8) > p:nth-child(1) > a')[0].text
        except :
            director = ""
            
        # 영화 장르
        try :
            try :
                genre = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.true > section > div.title.ott_true > div > span.genres')[0].text
            except :
                genre = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.false > section > div.title.ott_false > div > span.genres')[0].text                
        except :
            genre = ""
                
        # 국가 코드, 정규표현식
        try :
            try :
                country = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.true > section > div.title.ott_true > div > span.release')[0].text
                country = re.findall("[A-Z]+", country)
                country = country[0]
            except :
                country = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.false > section > div.title.ott_false > div > span.release')[0].text
                country = re.findall("[A-Z]+", country)
                country = country[0]
        except :    
            country = ""
            
        # 영상시간
        try :
            try :
                runtime = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.true > section > div.title.ott_true > div > span.runtime')[0].text 
            except :
                runtime = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.false > section > div.title.ott_false > div > span.runtime')[0].text
        except : 
            runtime = ""
                
        try :
            runtime = runtime.replace("1h", "60")
            runtime = runtime.replace("m","")
        except :
            pass
                    
        try :
            runtime = runtime.replace("2h", "120")
            runtime = runtime.replace("m","")
            runtime = int(runtime)
        except :
            pass

        try :
            if runtime[1] :
                runtime = runtime.split()
                hour = runtime[0]
                minute = runtime[1]
                runtime = int(hour) + int(minute)
            elif not runtime[1] :
                pass
        except :
            pass
                
        # 줄거리
        try :
            try :    
                overview = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.true > section > div.header_info > div > p')[0].text
            except :
                overview = driver.find_elements_by_css_selector('#original_header > div.header_poster_wrapper.false > section > div.header_info > div > p')[0].text
        except :
            overview = ""
                
                
        temp_id.append(site_id)
        temp_title.append(title)
        temp_release.append(release)
        temp_director.append(director)
        temp_url.append(url)
        temp_genre.append(genre)
        temp_country.append(country)
        temp_runtime.append(runtime)
        temp_overview.append(overview)
    
TMDB_DATA = pd.DataFrame({'TMDB_ID' : temp_id,
                        'TITLE' : temp_title, 
                        'RELEASE' : temp_release,
                        'GENRE' : temp_genre,
                        'DIRECTOR' : temp_director, 
                        'ACTOR' : None,
                        'COUNRTY' : temp_country,
                        'RUNTIME' : temp_runtime,
                        'PRODUCTION' : None,
                        'OVERVIEW' : temp_overview,
                        'URL' : temp_url,
                        })
    
print(time.time() - start)
# TMDB_DATA.to_excel('TMDB.xlsx', index=False)
    
    
print('------------------------------------------------------------------------------------------------')
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
driver.close()
```
